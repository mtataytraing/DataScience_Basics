2. Escribe un programa Pandas para contar la supervivencia por género, categorías y edad de varias clases. Siendo las categorías de edad: (0, 10), (10, 30), (30, 60), (60, 80)

```python
#supervivencia por clases; hay que hacer una "extensión de estos datos"
dfs=df.pivot(columns=['class'],values='survived') #hay que quitar los [] para que no me salga como subind.
dfs
```

```python
ndf=pd.concat([df,dfs],axis=1)
ndf
```

```python
ndf.groupby(['sex','rangos_edad']).agg({'First':'count','Second':'count','Third':'count'})
```
