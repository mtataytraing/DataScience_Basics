4. Escribe un programa Pandas para contar el mÃºmero de retrasos de salida por mes y origen.

```python
df.groupby(['month','origin']).first()
```

```python
import pandas as pd
```

```python
def contar_delay_positivo(columna_delay):
    contador=0
    for x in columna_delay:
        if x>0:
            contador=contador+1
    return contador
```

```python
df2=df.groupby(['month','origin']).agg(n_retrasos=('dep_delay',contar_delay_positivo))
df2
```

```python
df2['n_retrasos'].sum()
```
